```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('data/cagi5_df.csv')
df.head()
```

```python
df['AbsValue'] = np.abs(df['Value'])
```

```python
f9_df = df[df['regulatory_element']=='release_F9'].groupby(['Pos'])['AbsValue'].max()
```

```python
for enh in df['regulatory_element'].unique():
  fig = plt.figure()
  print(enh)
  enh_df = df[df['regulatory_element']==enh].groupby(['Pos'])['AbsValue'].max()
  plt.scatter(enh_df.index.values, enh_df.values)
  plt.show()
```

```python
for enh in df['regulatory_element'].unique():
  fig = plt.figure()
  print(enh)
  enh_df = df[df['regulatory_element']==enh]['Value']
  plt.hist(enh_df, bins=20, range=(-1.1, 1.1))
  plt.show()
```

```python
f9_df = df[df['regulatory_element']=='release_F9']['Value']
plt.hist(f9_df, bins=10, range=(-1.1, 1.1))
```
