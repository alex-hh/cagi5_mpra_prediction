```python
import pandas as pd
import numpy as np
```

```python
answerdf = pd.read_csv('data/cagi4_mpra/4-eQTL-causal_SNPs_answer_key2.txt', delimiter='\t')
answerdf.head()
```

```python
def get_prediction_df(group, which_set=2, method_num=1):
  grouppath = 'data/cagi4_mpra/4-eQTL-causal_SNPs_predictions/Group_{}/'.format(group)
  filepath = grouppath + 'Group_{}-prediction_file{}-{}.txt'.format(group, which_set, method_num)
  preddf = pd.read_csv(filepath, delimiter='\t')
  return preddf

group4_preds = get_prediction_df(4, 2, 1)
```

```python
group4_preds.head()
```

### emVar prediction task results

|Group|AUROC|AUPRC|
|----|-------|-------|
| 4 | 0.65526 | 0.45256

```python
from sklearn.metrics import roc_auc_score, average_precision_score

print(roc_auc_score(answerdf['emVar_Hit'], group4_preds['emVar_Hit']),
      average_precision_score(answerdf['emVar_Hit'], group4_preds['emVar_Hit']))

# group4's reported roc_auc_score
```
